### Load the Pre‑Trained YOLOv8 Model

```python
model = YOLO('yolo11n.pt')  # initialized with the pre-trained weights

## tell the model file will be automatically downloaded the first time you run this cell.
```

### make this interactive

```python
# Download a sample image to work with
!wget -O sample.jpg https://ultralytics.com/images/bus.jpg

```

```python
# Run object detection on the sample image
results = model("sample.jpg")
```

```python
# Display the image using matplotlib.
plt.figure(figsize=(10, 6))
plt.imshow(results[0].plot())
plt.show()
```

### Upload a Custom Image

```python
# pt to upload an image file ???from their computer.

## interactive output to add image

from google.colab import files
uploaded = files.upload()
```

```python
for fn in uploaded.keys():
    image_path = fn
    print(f'User uploaded file "{fn}" ({len(uploaded[fn])} bytes)')
    results = model(image_path, conf=0.5)

    # Display the detection result.
    plt.figure(figsize=(10, 6))
    plt.imshow(results[0].plot())
    plt.axis('off')
    plt.title("YOLOv11 Detection on Uploaded Image")
    plt.show()
```

```python
for fn in uploaded.keys():
    image_path = fn
    print(f'User uploaded file "{fn}" ({len(uploaded[fn])} bytes)')


    results = model(image_path, conf=0.8)

    plt.figure(figsize=(10, 6))
    plt.imshow(results[0].plot())
    plt.axis('off')
    plt.title("Detection with Confidence Threshold = 0.7")
    plt.show()
```
